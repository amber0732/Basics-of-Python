## Importing General Libraries of Pyhton :-

```python
import numpy as np
import pandas as pd 
```

```python
## Importing Data Visualization Library of Python :-
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Cust_Spend_Data.csv")
```

```python
df
```

## 1. Let's Draw a Histogram:-
## Histogram is created by using "matplotlib.pyplot Library".

```python
plt.hist(df["Avg_Mthly_Spend"])
plt.title("Average Montly Spend Distribution", fontsize = 20)
plt.xlabel("Money Spend", fontsize = 15)
plt.ylabel("No. of Months", fontsize = 15)
plt.show()
```

## 2. Let's Create a Boxplot graph :-
### Boxplot can be created only by using Seaborn Library.

```python
sns.boxplot(df["No_Of_Visits"], color = 'Red')
plt.title("Number of Visits Distribution", fontsize = 20, color = "Blue")
plt.xlabel("Frequency", fontsize = 15, color = "Blue")
plt.ylabel("No. of Visits", fontsize = 15, color = "Blue")
```

## 3. Let's Create a Bar Graph :-

```python
df["Apparel_Items"].value_counts().plot(kind = "bar", color = "Teal")
plt.title("Apparel Items Distribution", fontsize = 20, color = "Green")
plt.xlabel("Apperal Items", fontsize = 15, color = "Green")
plt.ylabel("Counts", fontsize = 15, color = "Green")
plt.show()
```

## 4. Let's Create a Horizontal Bar Graph :-

```python
df["FnV_Items"].value_counts().plot(kind = "barh", color = "Purple")
plt.title("Favourite Items Distribution", fontsize = 20, color = "Red")
plt.xlabel("Counts", fontsize = 15, color = "Red")
plt.ylabel("Items", fontsize = 15, color = "Red")
plt.show()
```

## 5. Now Let's a Pie Chart :-

```python
stp = df["Staples_Items"].value_counts()
keys = stp.keys().to_list()
counts = stp.to_list()
plt.pie(x = counts, labels = keys,  autopct = '%1.1f%%')
stp
plt.show()
```

## 6. Now Let's try the Scatter Dot graph :-

```python
plt.scatter(x = "Cust_ID", y = "Name", data = df)
plt.title("Cust_ID & Names Distribution")
plt.xlabel("Cust_ID")
plt.ylabel("Names")
plt.show()
```

```python
plt.scatter(x = "Avg_Mthly_Spend", y = "No_Of_Visits", data = df, color = "Orange")
plt.title("Average Monthly Spend & No. of Visits Distribution", fontsize = 17, color = "Teal")
plt.xlabel("Average Monthly Spend", fontsize = 13, color = "Teal")
plt.ylabel("No. of Visits", fontsize = 13, color = "Teal")
plt.show()
```

```python
plt.scatter(y = "Avg_Mthly_Spend", x = "No_Of_Visits", data = df, color = "Black")
plt.title("Average Monthly Spend & No. of Visits Distribution", fontsize = 17, color = "Teal")
plt.xlabel("No. of Visits", fontsize = 13, color = "Teal")
plt.ylabel("Average Monthly Spend", fontsize = 13, color = "Teal")
plt.show()
```

```python
plt.scatter(data = df, x = "FnV_Items", y = "FnV_Items", color = "Teal")
plt.title("Favorite Items Distribution", fontsize = 17, color = "Brown")
plt.xlabel("Favorite Items", fontsize = 13, color = "Brown")
plt.ylabel("Favorite Items", fontsize = 13, color = "Brown")
plt.show()
```
