## Note :- Best case to remove outliers is median in numerical data also with mean for moderate case. And for Catagorical data use mode to replace outliers.

## Let's Remove the Outiers :-

```python
df.boxplot("Age")
```

## Steps for Removing the Outliers :-

```python
def remove_outlier(col):
    sorted(col)
    q1,q3 = col.quantile([0.25, 0.75])
    iqr = q3 - q1
    lower_range = q1 - (1.5*iqr)
    upper_range = q3 + (1.5*iqr)
    return lower_range, upper_range
```

```python
lowage, uppage = remove_outlier(df["Age"])
df["Age"] = np.where(df["Age"]>uppage, uppage, df["Age"])
df["Age"] = np.where(df["Age"]<lowage, lowage, df["Age"])
```

## Outliers are Removed :-

```python
df.boxplot("Age")
```
